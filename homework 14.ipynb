```python
import pandas as pd
data = pd.read_csv('ratings.csv')
def rated(param):
    if param['rating'] <= 2:
        return 'низкий рейтинг'
    elif param['rating'] <4:
        return 'средний рейтинг'
    return 'высокий рейтинг '
def rate(df):
    df['rated'] = df.apply(rated, axis=1)
rate(data)
data.head(10)
```

```python
data = pd.read_csv('keywords.csv')
def region(string):
    splitted = string['keyword'].split(' ')
    for word in splitted :
        if word in('москва', 'тула', 'ярославль' ):
            return 'Центр'
        elif word in ('петербург', 'псков', 'мурманск'):
            return 'Северо-Запад'
        elif word in ('владивосток', 'сахалин', 'хабаровск'):
            return 'Дальний Восток'
    return 'undefined'
data['region'] = data.apply(region,axis =1)
data.head(130)
```

```python
import re
years = list(range(1950,2011))
dataset = pd.read_csv('ratings.csv')
data = pd.read_csv('movies.csv')
def production_year(data):
    k = re.findall(r'[\(](\d{4})[\)]' , data['title'])
    if k != []:
        if int(k[0]) in years:
            return int(k[0])
    return '1900'
data['year'] = data.apply(production_year, axis = 1)
data.merge(dataset,left_on='movieId', right_on='movieId', how='inner') \# перенос строки
.groupby(['movieId','year']).agg({'rating':'mean'}).sort_values('rating',ascending=False).head(20)
```
